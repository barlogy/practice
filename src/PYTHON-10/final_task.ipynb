### Успеваемость Студентов
    Проверьте гипотезу: у студентов с оплачиваемым питанием средний балл по математике выше, чем у студентов с льготным питанием.
    В качестве ответа напишите наибольший средний балл по математике среди этих групп студентов.

```python
mask_paid = student_data['lunch'] == 'standard'
mask_free = student_data['lunch'] == 'free/reduced'
mean_paid = student_data[mask_paid]['math score'].mean()
mean_free = student_data[mask_free]['math score'].mean()
print(round(max(mean_paid, mean_free)))
```

### Успеваемость Студентов
    Каков % студентов, родители которых имеют высшее образование уровня бакалавриата?

```python
res = student_data['parental level of education'].value_counts(normalize=True)['bachelor\'s degree'] * 100
print(round(res))
```

### Успеваемость Студентов
    Насколько медианный балл по письму у студентов в расовой группе А отличается от среднего балла по письму у студентов в расовой группе C?

```python
mask_a = student_data['race/ethnicity'] == 'group A'
median_a = student_data[mask_a]['writing score'].median()
mask_c = student_data['race/ethnicity'] == 'group C'
mean_c = student_data[mask_c]['writing score'].mean()
print(round(abs(median_a - mean_c)))
```
