### Недвижимость в Мельбурне
    Прочитать данные из CSV-файла в DataFrame.
    Преобразаовать даные первоначального DataFrame.
    Вывести информацию о DataFrame и первые строки таблицы.

```python
import pandas as pd

melb_data = pd.read_csv('data/melb_data_ph10.csv', sep=',')
melb_data['Car'] = melb_data['Car'].astype('int64')
melb_data['Bedroom'] = melb_data['Bedroom'].astype('int64')
melb_data['Bathroom'] = melb_data['Bathroom'].astype('int64')
melb_data['Propertycount'] = melb_data['Propertycount'].astype('int64')
melb_data['YearBuilt'] = melb_data['YearBuilt'].astype('int64')
melb_data.info()
melb_data.head(n=10)
```

### Недвижимость в Мельбурне
    Во сколько раз площадь прилегающей территории, с индексом 3521, больше площади участка с индексом 1690?

```python
print(round(melb_data.loc[3521, "Landsize"] / melb_data.loc[1690, "Landsize"]))
```

### Недвижимость в Мельбурне

### Недвижимость в Мельбурне
    Чему равно стандартное отклонение расстояния от центра города до объекта недвижимости?

### Недвижимость в Мельбурне
    Чему равно отклонение (в %) медианного значения площади здания от его среднего значения?

```python
building_median = melb_data['BuildingArea'].median()
building_mean = melb_data['BuildingArea'].mean()
print(round((abs(building_median - building_mean) / building_mean) * 100))
```

### Недвижимость в Мельбурне
    Сколько есть объектов недвижимости, стоимость которых составила больше 3 миллионов и были проданы риелторской компанией Nelson.

### Недвижимость в Мельбурне

### Недвижимость в Мельбурне
    Какова средняя цена объектов недвижимости с ценой менее одного миллиона, в которых либо количество комнат больше пяти, либо здание моложе 2015 года?

```python
mask = (melb_data['Price'] < 1_000_000) & ((melb_data['YearBuilt'] > 2015) | (melb_data['Rooms'] > 5))
print(round(melb_data[mask]['Price'].mean()))
```

### Недвижимость в Мельбурне
    В каком районе Мельбурна чаще всего продаются виллы и коттеджи (тип здания — h) с ценой меньше трёх миллионов?

```python
mask = (melb_data['Price'] < 3_000_000) & (melb_data['Type'] == 'h')
print(melb_data[mask]['Regionname'].mode()[0])
```
