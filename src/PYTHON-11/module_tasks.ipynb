### Недвижимость в Мельбурне
    Прочитать данные из CSV-файла в DataFrame и вывести информацию о DataFrame.
    Создать копию первоначального DataFrame.
    Вывести информацию о DataFrame и первые строки таблицы.

```python
import pandas as pd
melb_data = pd.read_csv('data/melb_data_ph11.csv', sep=',')
melb_df = melb_data.copy()
melb_data.info()
melb_data.head()
```

 ### Недвижимость в Мельбурне
    Удалим столбцы index и Coordinates из таблицы.

```python
# melb_df.drop(['index','Coordinates'],axis=1,inplace=True)
if 'index' in melb_df.columns and 'Coordinates' in melb_df.columns:
    melb_df = melb_df.drop(['index', 'Coordinates'], axis=1)
melb_df.head()
```

### Недвижимость в Мельбурне
    Создавать новые признаки.

```python
# MeanRoomsArea
total_rooms = melb_df['Rooms'] + melb_df['Bedroom'] + melb_df['Bathroom'] # общее количество комнат в здании
melb_df['MeanRoomsArea'] = melb_df['BuildingArea'] / total_rooms # средняя площадь одной комнаты для каждого объекта

# PricePerSqm
melb_df['PricePerSqm'] = melb_df['Price'] / melb_df['BuildingArea'] # цена за квадратный метр

# AreaRatio
diff_area = melb_df['BuildingArea'] - melb_df['Landsize'] # разница между площадью здания и площадью участка
sum_area = melb_df['BuildingArea'] + melb_df['Landsize'] # сумма площади здания и площади участка
melb_df['AreaRatio'] = diff_area/sum_area # отношение разницы площади здания и площади участка к их сумме
melb_df.head()
```

### Недвижимость в Мельбурне
    Преобразовывать столбцы с датами

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst=True)
display(melb_df['Date'])
```

### Недвижимость в Мельбурне
    За какой интервал времени (в годах) представлены данные, и на какой год приходится наибольшее число продаж?

```python
years_sold = melb_df['Date'].dt.year
print(years_sold)
print('Min year sold:', years_sold.min())
print('Max year sold:', years_sold.max())
print('Mode year sold:', years_sold.mode()[0])
```

### UFO

### Недвижимость в Мельбурне
    Напишите функцию get_weekend(weekday), которая принимает на вход элемент столбца WeekdaySale и возвращает 1, если день является выходным, и 0 — в противном случае.
    Создайте столбец Weekend, примените эту функцию к столбцу и вычислите среднюю цену объекта недвижимости, проданного в выходные дни.

```python
def get_weekend(weekday):
    if weekday in [5, 6]:
        return 1
    else:
        return 0

melb_df['Weekend'] = melb_df['WeekdaySale'].apply(get_weekend)
mean_price_weekend = melb_df[melb_df['Weekend'] == 1]['Price'].mean()
print(round(mean_price_weekend))
```

### Недвижимость в Мельбурне
    Преобразуйте столбец SellerG - оставьте в столбце только 49 самых популярных компаний, а остальные обозначьте как 'other'.
    Найдите, во сколько раз минимальная цена объектов недвижимости, проданных компанией 'Nelson', больше минимальной цены объектов, проданных компаниями, обозначенными как 'other'

```python
sellers_counts = melb_df['SellerG'].value_counts()
top_49_sellers = sellers_counts.nlargest(49).index
melb_df['SellerG'] = melb_df['SellerG'].apply(lambda x: x if x in top_49_sellers else 'other')
min_price_nelson = melb_df[melb_df['SellerG'] == 'Nelson']['Price'].min()
min_price_other = melb_df[melb_df['SellerG'] == 'other']['Price'].min()
ratio_min_prices = min_price_nelson / min_price_other
print(round(ratio_min_prices, 2))
```
