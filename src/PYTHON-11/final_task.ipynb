### Citibike
    Создайте признак времени суток поездки time_of_day.
    Время суток определяется следующим образом:
    - night (ночь) — с 00:00 до 06:00
    - morning (утро) — с 06:01 до 12:00
    - afternoon (день) — с 12:01 до 18:00
    - evening (вечер) — с 18:01 до 23:59
    Сколько поездок начиналось утром?
    Сколько поездок начиналось днём?
    Сколько поездок начиналось вечером?
    Сколько поездок начиналось ночью?

```python
def get_time_of_day(hour):
    try:
        # Преобразуем вход в число (поддерживает строки вида "12" или "12.0")
        h = int(float(hour))
    except (ValueError, TypeError):
        return None
    # Проверяем, что час в пределах 0–23
    if not 0 <= h <= 23:
        return None
    # Определяем время суток по диапазону
    if h <= 6:          # 0–6
        return "night"
    elif h <= 12:       # 7–12
        return "morning"
    elif h <= 18:       # 13–18
        return "afternoon"
    else:               # 19–23
        return "evening"

citibike_trip_cp['time_of_day'] = citibike_trip_cp['starttime'].dt.hour.apply(get_time_of_day)
num_morning_trips = citibike_trip_cp[citibike_trip_cp['time_of_day'] == 'morning'].shape[0]
num_afternoon_trips = citibike_trip_cp[citibike_trip_cp['time_of_day'] == 'afternoon'].shape[0]
num_evening_trips = citibike_trip_cp[citibike_trip_cp['time_of_day'] == 'evening'].shape[0]
num_night_trips = citibike_trip_cp[citibike_trip_cp['time_of_day'] == 'night'].shape[0]
print('Number of trips that started in the morning/afternoon/evening/night:', num_morning_trips, num_afternoon_trips, num_evening_trips, num_night_trips)
```

### Citibike
    Во сколько раз количество поездок, совершённых днём, больше, чем количество поездок, совёршенных ночью, за представленный в данных период времени?

```python
mask_afternoon = citibike_trip_cp['time_of_day'] == 'afternoon'
mask_night = citibike_trip_cp['time_of_day'] == 'night'
num_afternoon_trips = citibike_trip_cp[mask_afternoon].shape[0]
num_night_trips = citibike_trip_cp[mask_night].shape[0]
ratio_afternoon_night = num_afternoon_trips / num_night_trips if num_night_trips != 0 else float('inf')
print('Ratio of afternoon trips to night trips:', round(ratio_afternoon_night))
```
